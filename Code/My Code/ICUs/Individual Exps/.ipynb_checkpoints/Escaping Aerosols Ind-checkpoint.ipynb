# ICU 1 Neutral Pressure 1

```python
ins = [0]*len(DO1_1['Time'])
doors = [0]*len(DO1_1['Time'])
ns = [0]*len(DO1_1['Time'])
totavg = [0]*len(DO1_1['Time'])


for x in range(len(DO1_1['Time'])):
    
    ins[x] = (((DO1_1['S-01'].values[x]+DO1_1['S-02'].values[x]+DO1_1['S-03'].values[x]+DO1_1['S-04'].values[x]+DO1_1['S-05'].values[x]+DO1_1['S-06'].values[x]+DO1_1['S-07'].values[x]+DO1_1['S-08'].values[x]+DO1_1['S-09'].values[x]+DO1_1['S-10'].values[x]+DO1_1['S-12'].values[x]+DO1_1['S-13'].values[x]+DO1_1['S-14'].values[x])/(100*180))) - (0.0065*13)
    #print("Inside =", inside[x])

    doors[x] = (((DO1_1['S-15'].values[x]+DO1_1['S-18'].values[x]+DO1_1['S-19'].values[x])))/(100*180) - (0.0065*3)
    #print("Door Sensors =", door_sensors[x])

    ns[x] = (DO1_1['S-16'].values[x]/(100*180)) - 0.0065
    #print("Nurse's Station =", nurses_station[x])
    
    totavg[x] = (((DO1_1['S-01'].values[x] + DO1_1['S-02'].values[x] + DO1_1['S-03'].values[x] + DO1_1['S-04'].values[x] + DO1_1['S-05'].values[x] + DO1_1['S-06'].values[x] + DO1_1['S-07'].values[x] + DO1_1['S-08'].values[x] + DO1_1['S-09'].values[x] + DO1_1['S-10'].values[x] + DO1_1['S-12'].values[x] + DO1_1['S-13'].values[x] + DO1_1['S-14'].values[x] + DO1_1['S-15'].values[x] + DO1_1['S-16'].values[x] + DO1_1['S-18'].values[x] + DO1_1['S-19'].values[x])/(100*180))) - (0.0065*17)
    #print("Total =", total[x])
    #print(x, '=', nurses_station[x]+inside[x]+door_sensors[x])   
    
averages = []
for x in range(200,400):
#     print("Inside fraction at ",x, "=", ins[x]/totavg[x])
#     print("Outside fraction at ",x, "=", doors[x]/totavg[x])
#     print("Nurse's Station fraction at ",x, "=", ns[x]/totavg[x])
#    print("Total Average at",x, "=", totavg[x])
    averages.append(totavg[x])
    
#print(averages)
#print(statistics.variance(averages))
m = statistics.mean(averages)
#print((statistics.variance(averages))/m)
print("percent difference = ", 1 - m/max(averages))

in_avg = 0
door_avg = 0
ns_avg = 0
tot_avg = 0

for x in range(200,400):
    in_avg = ins[x]/totavg[x] + in_avg
    door_avg = doors[x]/totavg[x] + door_avg
    ns_avg = ns[x]/totavg[x] + ns_avg
    tot_avg = totavg[x]/totavg[x] + tot_avg

print("ns =", ns_avg/200)
print("in_avg =", in_avg/200)
print("door_avg =", door_avg/200)
print("total_avg =", tot_avg/200)

fig = plt.figure(1, figsize=(12,8), dpi=64)
plt.semilogy(DO1_1['Time'], ins, linewidth=5.0)
plt.semilogy(DO1_1['Time'], doors, linewidth=5.0)
plt.semilogy(DO1_1['Time'], ns, linewidth=5.0)
plt.ylim([0.0003, 1])
plt.title('ICU 1 Neutral Pressure 1').set_size(25)
plt.xlabel('Time (s)').set_size(25)
plt.ylabel('Normalized Particle Concentration').set_size(25)
axes = plt.gca()
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.axhline(y=0.0065, color='black', linestyle='--', linewidth=5.0)
plt.legend(['Zone 1', 'Zone 2', 'Zone 3', 'Background'], prop = {'size':20})
```
